# Distributed Statistical Inference and Learning

# Inference under information constraints : Testing Private Coin Protocol

Abstract
Multiple players are each given one independent sample, about which they can only provide limited information to a central referee. Each player is allowed to describe its observed sample to the referee using a channel from a family of channels W, which can be instantiated to capture,among others, both the communication- and privacy-constrained settings. The referee uses the players’ messages to solve an inference problem on the unknown distribution that generated the samples. We derive lower bounds for the sample complexity of learning and testing discrete distributions in this information constrained setting. Underlying our bounds is a characterization of the contraction in chi-square distance between the observed distributions of the samples when information constraints are placed. This contraction is captured in a local neighborhood in terms of chi-square and decoupled chi-square fluctuations of a given channel, two quantities we introduce. The former captures the average distance between distributions of channel output for two product distributions on the input, and the latter for a product distribution and a mixture of product distribution on the input. Our bounds are tight for both public- and privatecoin protocols. Interestingly, the sample complexity of testing is order-wise higher when restricted to private-coin protocols.

Reference : Inference under information constraints I: Lower bounds from chi-square contraction. Jayadev Acharya; Clément L. Canonne; and Himanshu Tyagi. IEEE Trans. Inform. Theory, 66(12): 7835–7855. 2020. Preprint available at arXiv:abs/1812.11476.

```python
# We will generate unifrom random samples from a single discrete distribution first then play around
# Let the parent distribution be multinomially distributed with probability 0.1,0.2,0.3,0.4 on 1,2,3,4

import math
import numpy as np

sample = np.random.multinomial(20, [0.1]+[0.2]+[0.3]+[0.4], 5)
sample

```

```python
# The point estimate = [0.1 0.2 0.3 0.4]*[20]I_4
# Then drop NaNs for lack of a better word

estimate = [2,4,6,8]
```

```python
import matplotlib as plt
from matplotlib.pyplot import hist

hist(sample)
```

```python
diff = []
for i in sample:
    diff.append(estimate - i)
```

```python
diff
```

```python
(sum(diff))
```
